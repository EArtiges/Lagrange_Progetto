```python
def get_coordinates(address):
    time.sleep(1.5)
    address=re.sub(r'S\.','San ',address)
    address=re.sub(r'Fr\.', '',address)
    address=re.sub(r'[A-Z]\.','',address)
    if not 'Mirafiori' in address:
        address=re.sub(r'Di','',address)
    if address=='8 Corso Siccardi':
        address='8 Corso Giuseppe Siccardi'
    try:
        url = 'https://nominatim.openstreetmap.org/search?q={},+torino&format=json&polygon=1&addressdetails=1'.format('+'.join(address.split(' ')))
        response = requests.get(url)
        soup=response.json()
        return soup[0]['lat'],soup[0]['lon']
    except IndexError:
        url = 'https://nominatim.openstreetmap.org/search?q={}&format=json&polygon=1&addressdetails=1'.format('+'.join(address.split(' ')))
        response = requests.get(url)
        soup=response.json()
        try:
            return soup[0]['lat'],soup[0]['lon']
        except IndexError:
            return ''
```

```python
lang = detect("Ein, zwei, drei, vier")
# it for italian, en for english
class StemmedCountVectorizer(CountVectorizer):
    # This class is basically a CountVectorizer class that has been enhanced with a stemmer method.
    def build_analyzer(self):
        #Use super to access the method build_analyzer form the class CountVectorizer.
        analyzer = super(StemmedCountVectorizer, self).build_analyzer()
        #return lambda doc: ([italian_stemmer.stem(w) for w in analyzer(doc)])
        return lambda doc: ([self.stemmer(w) for w in analyzer(doc)])
    
    def stemmer(self, w):
        if langid.classify(w)=='it':
            return italian.stemmer(w)
        elif langid.classify(w)=='en':
            return english.stemmer(w)
        else:
            return w
        
    def language_detector(w):
        try:
            return langid.classify(w)[0]
        except LangDetectException:
            return 'emoji'
    
```
