```python
df_new=pd.read_csv('Torino_data/torino_posts.csv',lineterminator='\n')
df_new=df_new.rename(columns = {'caption':'text','taken_at_timestamp':'date'})
```

```python
plt.rcParams["figure.figsize"] = 12,4
plt.style.use('ggplot')
```

```python
plt.figure()
for year in range(2010,2019):
    df_2018=df_new[df_new.date>'{}-01-01 00:00:00'.format(year)]
    df_2018=df_2018[df_2018.date<'{}-01-01 00:00:00'.format(year+1)]
    timestamps=df_2018.date.tolist()
    timestamps=[t.split(' ')[1] for t in timestamps]
    clock=[':'.join([str(m/60), str(m%60), '00']) for m in range(0,24*60,5)]
    Histo=np.histogram([int(i)*60+int(j) for i,j in [[t.split(':')[0:2]][0] for t in timestamps]],144)
    plt.plot(Histo[1][:-1],Histo[0])
plt.legend(range(2010,2019))
plt.title('daily tweet distribution for each year')
plt.savefig('daily_tweet_distrib.pdf')
plt.show()
```
