```python
def normalized(v):
    Norm=np.sqrt(sum([u**2 for u in v]))
    return ([float(u)/Norm for u in v])
```

```python
val_dates=df_valentino.date.tolist()
all_dates=df_new.date.tolist()
years=set([d.year for d in val_dates])
```

```python
# Count the number of posts that were generated in parco del valentino.
# To account for the evolution of the dataset, compute the proportion of the dataset for each year that was
# created in the parco del valentino.
# To find out whether or not some part of the day became more prominent along the time,
# normalize the vector and plot it.

n_bins=48

plt.figure(figsize=(8,8))
for year in sorted(years)[3:]:
    print(year)
    times=[t.hour*60 + t.minute for t in val_dates if t.year==year]
    all_times=[t.hour*60 + t.minute for t in all_dates if t.year==year]
    val_histo_time=np.histogram(times,bins=n_bins)
    all_histo_time=np.histogram(all_times,bins=val_histo_time[1])
    plt.plot([datetime.time(int(h)/60,int(h)%60,0) for h in val_histo_time[1]][:-1],normalized([float(val_histo_time[0][i])/all_histo_time[0][i] for i in range(n_bins)]))
#    plt.plot([datetime.time(int(h)/60,int(h)%60,0) for h in val_histo_time[1]][:-1],[float(val_histo_time[0][i])/all_histo_time[0][i] for i in range(n_bins)])

plt.legend(sorted(years)[3:])
plt.title('Normalized proportion of posts in the parco del valentino per year')
plt.savefig('Norm_prop_N_instaposts_val.png')
plt.show()
```
