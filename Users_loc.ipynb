```python
#DF/USERS holds the same users as USERS bt without doubles.
DFUSERS=os.listdir('Twitter_scraping/DF/USERS')
```

```python
from urllib.error import HTTPError
from twitter import TwitterHTTPError
import time
#For all users having mentioned that they are based in Torino:
followers_torinese=[]
scraped_users=[''.join([c for c in w if c.isdigit()]) for w in DFUSERS]

for r in scraped_users:
    user_id=r #['id_str']
    next_cursor=-1
    followers=[]
    while(True):
        #We use the cursoring method of Twitter's API. 
        try:
            query = twitter.followers.ids(user_id=user_id, cursor=next_cursor)
        except (HTTPError, TwitterHTTPError):
            time.sleep(60)
        next_cursor= query['next_cursor']
        #We add the current result to out list and continue
        followers+=query
        if next_cursor == 0:
            #we reached the end of this user timeline
            break
        
    #list of all str(id) of followers of user_id=r. We want to learn more about them: are they from Torino?
    ids_string=[str(u) for u in query['ids']]
    subquery=[]
    #We can't ask for more than 100 at a time
    while(ids_string):
        #Let's take them all if it's less than 100; else, just the first 100
        users=','.join(ids_string[0:min(len(ids_string),100)])
        del ids_string[0:min(len(ids_string),100)]
        subquery += twitter.users.lookup(user_id = users)
    for user in subquery:
        if ('Torino' or 'Turin') in user['location']:
            followers_torinese.append(user['id_str'])
            #Collect the user only if s.he is from Torino
print('{} unique users have mentioned Torino or Turin in their location'.format(len(set(followers_torinese))))
```

```python
f=open('followers_torinese_2.txt','w')
for i in set(followers_torinese):
    if i not in scraped_users:
        f.write(i + ' \n')
f.close()
```
