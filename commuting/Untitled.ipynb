```python
o_d_n = [ i for i in zip(df_transport['name_o'],df_transport['name_d'],df_transport['spostamenti'])]
for o in set([i[0] for i in o_d_n if 'TORINO' in i[0]]):
    dest=[]
    amount=[]
    for element in o_d_n:
        if element[0]==o and 'TORINO' in element[1]:
            dest.append(element[1])
            amount.append(element[2])
    plt.figure()
    plt.plot([i for i in range(len(amount))],sorted(amount)[::-1])
   # plt.xlabel(dest)
    plt.title(o)
    plt.savefig(o+'.pdf')
    plt.show()
    plt.close()
```

```python
#The file we got from the TOrino website is in Gauss-Boaga projection. its EPSG code is 3003 but it does not know it.
df_ngh=geopd.read_file('neighbourhoods/distretti_assistenziali_geo/distretti_assistenziali_geo.shx')
#We have to tell it:
df_ngh.crs={'init':'epsg:3003'}
#And then to plot it in the usual GPS coordinates.
df_ngh.to_crs(epsg=4326).plot(figsize=(8,8), cmap='Blues')#.to_crs({'proj': 'merc'}).plot()
plt.axis('off')
```

```python
df_ngh.CIRCOSCRIZ=[str(e) for e in df_ngh.CIRCOSCRIZ]
```

```python
df_ngh['INDIC']=df_ngh['CIRCOSCRIZ']+df_ngh['LETTERA']
```

```python
df_ngh.drop(['ID_DISTR','LETTERA'], axis=1,inplace=True)
```

```python
from shapely.ops import cascaded_union
```

```python
df_ngh.plot()
```

```python
df_ngh['centers']=df_ngh.groupby('CIRCOSCRIZ').geometry.apply(lambda x:x.centroid)
```

```python
df_ngh
```

```python
code_centers_dict={}
for cd, ct in zip(df_ngh.INDIC, df_ngh.centers):
    code_centers_dict[cd]=ct
```

```python
mask=['TORINO' in e for e in df_transport['name_o']]
maskb=['TORINO' in e for e in df_transport['name_d']]
mask=[mask[i] and maskb[i] for i in range(0, len(mask))]
```

```python
df_transport=df_transport[mask]
```

```python
df_transport.name_o=df_transport.name_o.apply(lambda x : x[9:].strip(' '))
```

```python
df_transport.name_d=df_transport.name_d.apply(lambda x : x[9:].strip(' '))
```

```python
df_transport
```

```python
loc_dest_dict={}
for key in dest_loc_dict.keys():
    V=dest_loc_dict[key]
    loc_dest_dict[V]=key
```

```python
loc_dest_dict
```

```python
code_centers_dict
```

```python
df_transport
```

```python
df_transport=df_transport[df_transport.name_o != 'RIVALTA DI TORINO']
df_transport=df_transport[df_transport.name_d != 'RIVALTA DI TORINO']
```

```python
loc_dest_dict
```

```python
df_transport['GPS_o']=df_transport.name_o.apply(lambda x : loc_dest_dict[x])
df_transport['GPS_o']=df_transport.GPS_o.apply(lambda x : code_centers_dict[x])

df_transport['GPS_d']=df_transport.name_d.apply(lambda x : loc_dest_dict[x])
df_transport['GPS_d']=df_transport.GPS_d.apply(lambda x : code_centers_dict[x])
```

```python
df_transport=df_transport[['GPS_o','GPS_d','spostamenti']]
```

```python
df_transport['geometry']=[z for z in zip(df_transport.GPS_o,df_transport.GPS_d)]
```

```python
from shapely.geometry import LineString
df_transport['geometry']=df_transport.geometry.apply(lambda x : LineString(x))
```

```python
geo_transport=geopd.GeoDataFrame(df_transport)
```

```python
geo_transport.crs={'init':'epsg:3003'}
#And then to plot it in the usual GPS coordinates.
ax=df_ngh.to_crs(epsg=4326).plot(figsize=(8,8), edgecolor='black',linewidth=1,color=(0,0,0,0))
df_centers=geopd.GeoDataFrame(df_ngh.centers)
df_centers['geometry']=df_centers.centers
df_centers.crs={'init':'epsg:3003'}
df_centers.to_crs(epsg=4326).plot(ax=ax,color='orange')
geo_transport.to_crs(epsg=4326).plot(figsize=(8,8), column='spostamenti',ax=ax, cmap='Blues',alpha=.2)
plt.savefig('commuting.pdf')
#.to_crs({'proj': 'merc'}).plot()
```
